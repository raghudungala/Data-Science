```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path = "C:\\Users\\dragh\\Downloads\\usage_data.csv"
```

```python
data = pd.read_csv(file_path)
```

```python
data.head()
```

```python

sessions_mean = data['Sessions'].mean()

sessions_median = data['Sessions'].median()
sessions_max = data['Sessions'].max()
sessions_min = data['Sessions'].min()
```

```python
time_spent_mean = data['Time spent'].mean()
time_spent_median = data['Time spent'].median()
time_spent_max = data['Time spent'].max()
time_spent_min = data['Time spent'].min()
```

```python
print(f"Sessions Mean: {sessions_mean}")
print(f"Sessions Median: {sessions_median}")
print(f"Sessions Max: {sessions_max}")
print(f"Sessions Min: {sessions_min}")

print(f"Time Spent Mean: {time_spent_mean}")
print(f"Time Spent Median: {time_spent_median}")
print(f"Time Spent Max: {time_spent_max}")
print(f"Time Spent Min: {time_spent_min}")
```

```python
data['Date'] = pd.to_datetime(data['Date'])
```

```python
plt.figure(figsize=(10, 6))
plt.plot(data['Date'], data['Sessions'], marker='o')
plt.title('Sessions Over Time')
plt.xlabel('Date')
plt.ylabel('Sessions')
plt.grid(True)
plt.show()
```

```python
avg_time_spent = data.groupby('Feature')['Time spent'].mean().sort_values()

# Plotting a bar chart
plt.figure(figsize=(10, 6))
avg_time_spent.plot(kind='bar', color='skyblue')
plt.title('Average Time Spent per Feature')
plt.xlabel('Feature')
plt.ylabel('Average Time Spent')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
# Plotting a scatter plot of Sessions vs. Time Spent
plt.figure(figsize=(8, 6))
plt.scatter(data['Sessions'], data['Time spent'], color='green', alpha=0.7)
plt.title('Sessions vs. Time Spent')
plt.xlabel('Sessions')
plt.ylabel('Time Spent')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Boxplot showing distribution of Time Spent across Features
plt.figure(figsize=(10, 6))
plt.boxplot([data[data['Feature'] == feature]['Time spent'] for feature in data['Feature'].unique()], 
            labels=data['Feature'].unique())
plt.title('Distribution of Time Spent across Features')
plt.xlabel('Feature')
plt.ylabel('Time Spent')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```
